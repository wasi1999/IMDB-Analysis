```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
imdb=pd.read_csv('IMDB-Movie-Data.csv',)
imdb.head()
```

```python
imdb.shape
```

### Have a look that whic column contains null value 

```python
imdb.isnull().any()
```

### Dropping those rows which contain null values 

```python
imdb=imdb.dropna(axis=0)
```

### In which year most films released?

```python
plt.figure(figsize=(15,6))
sns.countplot(imdb['Year'],palette='mako_r')
plt.xlabel('Year',fontsize=20)
plt.ylabel('No.Of Movies',fontsize=20)
```

```python

imdb['Rating'].head(10)
```

```python
some_rating=(8.1,7.0,7.3,7.2,8.3,7.1,7.0,7.5)
top_rating=imdb.loc[imdb['Rating'].sort_values(ascending=False).isin(some_rating)]
```

```python
top_40=imdb.sort_values(by='Rating',ascending=False).head(40)
top_40
```

```python
plt.figure(figsize=(15,7))
sns.barplot(top_40['Title'],top_40['Rating'])
plt.title('Top 40 Movies',fontsize=25)
plt.xlabel('Title',fontsize=20)
plt.ylabel('Rating',fontsize=20)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(15,7))
plt.hist(top_40['Rating'])
```

```python
plt.figure(figsize=(19,7))
sns.countplot(imdb['Rating'])
plt.title('Counts Of Rating',fontsize=25)
plt.xlabel('Rating',fontsize=20)
plt.ylabel('No.Of Counts',fontsize=20)
```

### Top 10 Directors with most votes

```python

dir_votes=imdb.groupby(['Director']).mean()[['Votes','Revenue (Millions)']].reset_index().head(10)
dir_votes['Votes']=dir_votes['Votes'].astype(int)
```

```python
dir_votes.sort_values(by='Votes',ascending=False).reset_index()
```

```python
plt.figure(figsize=(19,10))
sns.barplot(dir_votes['Director'],dir_votes['Votes'],palette='magma')
plt.xticks(rotation=30)
plt.title('Top 10 Directors On The Basis Of Votes',fontsize=25)
plt.xlabel('Directors',fontsize=20)
plt.ylabel('Votes',fontsize=20)
plt.show()
```
